```python
import os
import csv
import pandas as pd
import re
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore') # 경고 무시
# from google.colab import drive
# drive.mount('./gdrive', force_remount=False)
plt.rc('axes', unicode_minus=False) # 마이너스 출력

# 파일 불러와서 데이터프레임에 저장
f = open('./gdrive/MyDrive/Colab Notebooks/temperature.csv', 'r', encoding='cp949')
data = csv.reader(f, delimiter=',')
df = pd.DataFrame(data).dropna()
f.close()

# 결측치가 발생해서 정규식으로 처리함
df = df.rename(columns=df.iloc[0]).drop(7)
df['최고기온(℃)'] = [float(*re.findall('-?\d+.\d+', item)) for item in df['최고기온(℃)']]

print(f'기상 관측 이래, 최고 기온이 가장 높았던 날은', df.loc[df['최고기온(℃)'].idxmax(), '날짜'], '로,', df['최고기온(℃)'].max(), '도 입니다.')
```

1950년도엔 전쟁, 그 이외의 경우엔 알 수 없는 이유로 결측치가 발생하였습니다.    
따라서 여러 결측치를 효과적으로 제거하기 위해, 정규식을 지원하는 파이썬의 내장모듈인 re모듈을 활용하였습니다.

```python
a = [str(re('\d+-0?1-0?1', item)) for item in df['날짜'] if str(re.findall('\d+-0?1-0?1', item)) != '[]']
print(a)
temp = pd.DataFrame(a, df[a])

plt.plot(df['날짜'], df['최고기온(℃)'])
plt.show()
```
