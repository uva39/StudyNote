1. matplotlib으로 이미지를 불러오기    
2. 이미지를 블러처리하기    
3. 이미지의 테두리를 따기

```python
import matplotlib.pyplot as plt
import numpy as np

imgr = plt.imread("check.png") # ".png" ".bmp" ".jpg"
img = imgr.copy()
```

```python
n = 2
f = np.ones([n, n]) / (n*n)
for i in range(img.shape[0]-(n-1)):
    for j in range(img.shape[1]-(n-1)):
        for k in range(img.shape[2]): # RGB 3채널에 대해 반복
            img[i, j, k] = np.sum(img[i:i+n, j:j+n, k] * f)

plt.imshow(img)
```

```python
import matplotlib.pyplot as plt
import numpy as np

imgr = plt.imread("check.png") # ".png" ".bmp" ".jpg"
img = imgr.copy()
```

```python
f = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])
for i in range(img.shape[0]-2):
    for j in range(img.shape[1]-2):
        for k in range(img.shape[2]):
            img[i, j, k] = np.abs(np.sum(img[i:i+3, j:j+3, k] * f))

img = img / img.max()
plt.imshow(img)
```
